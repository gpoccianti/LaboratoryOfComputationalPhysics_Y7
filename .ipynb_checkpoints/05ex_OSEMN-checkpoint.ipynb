1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
def random_file_w(N, file_name):
    # Generate N random numbers
    random_numbers = np.random.rand(N)

    # Write these numbers to a file
    with open(file_name, mode='w') as f:
        for number in random_numbers:
            f.write(f"{number}\n")  # Write each number on a new line
            
def random_file_r(N,file_name):
    # Open the file in read mode to read and print its contents
    with open(file_name, mode='r') as f:
        for line in f:
            print(line.strip())  # Print each line (remove trailing newline)
            for c in line.split():  # Split and demonstrate split() method
                print(c)
```

```python
# Example usage
random_file_w(5, "simple_data.txt")
random_file_r(5, "simple_data.txt")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
#writes the matrix into the file_name (.txt) file
def vec_file_w(mat, file_name):

    # Write these numbers to a file
    with open(file_name, mode='w') as f:
        for row in mat:
            for xx in row:
                f.write(f"{xx} ")  #Write each number
            f.write("\n")#Write a newline after each column


#Another version (without the final spaces):
def vec_file_w(mat, file_name):

    # Write these numbers to a file
    with open(file_name, mode='w') as f:
        f.write(str(mat))
 #       for row in mat:
 #           f.write(" ".join(map(str, row)))  # Join numbers with spaces and write them
 #           f.write("\n")  # Write a newline after each row
```

```python
N=5
M=5

mat=np.random.rand(M,N)

vec_file_w(mat,"data.txt")

!cat data.txt
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
#We manually modify the file to a .csv by subsituting the spaces to commas
with open("data.txt", mode='r') as f:
    lines=f.readlines()
    # Replace spaces with commas in each line
    lines = [line.replace("[","").replace(" ]","").replace("]","").strip().replace(" ", ",") + "\n" for line in lines]
with open("data.csv", mode='w') as f:
    f.writelines(lines)
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
file_credit_card= "credit_card.dat"
with open(file_credit_card, mode='r') as f4:
    for line in f4:
        #print(line)
        no_space_line = line.strip()[:-4] ##exclude last 4 bits (spaces) f each line
        credit_card_number = ""
        for i in range(0, len(no_space_line),6): ##in each line iterate every 6 every bits
            credit_card_number += chr(int(no_space_line[i:i+6],2)) # Convert 6-bits binary number to decimal, then to character
        print(credit_card_number)
```

```python
#print(df.head())
print("List of labels: ", df.columns.tolist()) #list of labels

#Now we print the first elements
print("\n")
print("First elements: \n")
print(df.head())
```

```python
#Draw the histogram of the class field:
#We create a function to do that
def hist_df(df,field: str,normalize=True):
    dg= df[[field]]
    # Plot the histogram
    plt.hist(dg[field], bins=10, edgecolor='black',density=normalize)  # Adjust `bins` as needed
    plt.title(f"Histogram of {field}")
    plt.xlabel(field)
    plt.ylabel("Occurrences")
    plt.show()

hist_df(df,"ring-number",normalize=False)
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:
    
  + features_1
  + features_2
  + features_3

```python
url = "https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1"
dg = pd.read_csv(url)
```

```python
df1=dg.iloc[:,1:4] #we slice df with the iloc method
print(df1.head())
```

```python
#We define a function to do the scatter plot:
def scatter_plot(df, feat_x: str,feat_y: str, x_label="Feature X", y_label="Feature Y", title="Scatter Plot"):
    """
    Create a scatter plot for the given x and y features.

    Parameters:

        df: Our dataset.
        feat_x (str): Label corresponding to the x-axis.
        feat_y (str): Label corresponding to the y-axis.
        x_label (str): Label for the x-axis (default is 'Feature X').
        y_label (str): Label for the y-axis (default is 'Feature Y').
        title (str): Title of the scatter plot (default is 'Scatter Plot').
    """
    plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
    plt.scatter(df[[feat_x]], df[[feat_y]], alpha=0.7, edgecolor='k')  # Scatter plot with transparency and edges
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)  # Add a grid for better visualization
    plt.show()

#We now create an array of features
labels=df1.columns.tolist()
#Generate all permutations of 2 elements
permutations=list(itertools.permutations(labels,2))
#Convert into a numpy array (optional)
perm_arr=np.array(permutations)

#Loop over the permutations
for per in perm_arr:
    scatter_plot(df1,per[0], per[1],str(per[0]),str(per[1]))
```
