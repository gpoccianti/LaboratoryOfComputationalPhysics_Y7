2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
#Find the outer product of:
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
#a) Using the function outer in numpy
w=np.outer(u,v)
print("Outer product (method 1): \n", w)

#b) Using a nester for loop or list comprehension
def ext_prod(u,v):
    M=u.size
    N=v.size
    w=np.zeros((M,N))
    w=[[u[i]*v[j] for j in np.arange(N)] for i in np.arange(M)]
    return np.array(w)

w2=ext_prod(u,v)
print("Outer product (method 2): \n", w2)

#c) Using numpy broadcasting operations
def ext_prod_broad(u,v):
    a=np.tile((v),(v.size,1))
    b=np.tile((u),(u.size,1)).T
    return a*b

w3=ext_prod(u,v)
print("Outer product (method 3): \n", w3)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)

````python
import numpy.random as npr
import matplotlib.pyplot as plt
N=1000
nsteps=200

def step(N,t): return np.ones((N,t))-2*npr.randint(2,size=(N,t)) #all the steps until time t

pos=np.sum(step(N,nsteps),axis=1) #pos at time t
pos**2
mean=np.mean(pos**2)
def msd(N,t):
    pos=np.sum(step(N,t),axis=1)
    return np.mean(pos**2)

y=list(map(lambda t: msd(N,t),np.arange(nsteps)))

plt.plot(np.arange(nsteps),y)
````

````python
import numpy.random as npr
from tqdm import tqdm
import matplotlib.pyplot as plt

#single random walker
#single step
def random_walk(pp=0.5, L=1000, T=1000, x0=0):
    '''
    Random walk with probability of going up pp and down 1-pp
    L: number of walkers
    T: number of steps
    x0: initial position
    '''
    # initialize
    x = np.zeros((L, T))
    x[:, 0] = x0
    # random walk
    for t in tqdm(range(1, T)):
        # random numbers
        r = np.random.rand(L)
        # update position
        x[:, t] = x[:, t-1] + (r < pp)*2 - 1
    return x
````

````python
#a) Take 1000 walkers and let them walk for 200 steps
npr.seed(123) # fix seed for reproducible result
L=1000 #walkers
steps= 200 #steps

wlkr=random_walk(0.5,L,steps)


#b) Use randint to create a 2D array of size walkers x steps with values -1 or 1
npr.randint(-1,1,L*steps)

#c) Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
dist=wlkr.sum(axis=0)
#d) Take the square of that 2D array (elementwise)
dist2=dist**2
print("Walking distances for each walker: \n", dist2)
#e) Compute the mean of the squared distances at each step (i.e. the mean along the columns)
dist_col=wlkr.sum(axis=1)**2
dist_col_mean=dist_col.mean()
print("Mean of the squared distances at each step: ", dist_col_mean)
````

````python
#Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
plt.plot(np.arange(steps),np.sqrt(dist2))
plt.plot(np.arange(steps),np.arange(steps))
plt.xlabel("t")
plt.ylabel(r"$ \sqrt{d^2}$")
plt.show()
````
